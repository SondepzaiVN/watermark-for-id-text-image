# 03 - Attack And Recovery Comparison
Notebook nay dung de so sanh anh sau tan cong va watermark duoc khoi phuc.

```python
from pathlib import Path
import cv2
import matplotlib.pyplot as plt

ROOT = Path('..').resolve()
results_dir = ROOT / 'results'

host_dirs = sorted([p for p in results_dir.iterdir() if p.is_dir() and (p / 'attack').exists() and (p / 'recover').exists()]) if results_dir.exists() else []
if not host_dirs:
    raise FileNotFoundError('Khong tim thay thu muc ket qua host nao trong results/.')

print('Danh sach host da co ket qua:')
for p in host_dirs[:20]:
    print(' -', p.name)
```

```python
selected_host = host_dirs[0]
attack_dir = selected_host / 'attack'
recover_dir = selected_host / 'recover'

attack_images = sorted([p for p in attack_dir.glob('*') if p.suffix.lower() in {'.png', '.jpg', '.jpeg', '.bmp'}])
recover_images = sorted([p for p in recover_dir.glob('*') if p.suffix.lower() in {'.png', '.jpg', '.jpeg', '.bmp'}])

print('Selected host:', selected_host.name)
print('Attack images:', len(attack_images))
print('Recover images:', len(recover_images))
```

```python
def read_rgb(path: Path):
    img = cv2.imread(str(path), cv2.IMREAD_COLOR)
    if img is None:
        return None
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

n_pairs = min(6, max(len(attack_images), len(recover_images)))
if n_pairs == 0:
    print('Host nay chua co anh attack/recover de hien thi.')
else:
    plt.figure(figsize=(12, 3 * n_pairs))
    for i in range(n_pairs):
        plt.subplot(n_pairs, 2, i * 2 + 1)
        if i < len(attack_images):
            img_a = read_rgb(attack_images[i])
            if img_a is not None:
                plt.imshow(img_a)
                plt.title(f'Attack: {attack_images[i].name}')
            else:
                plt.text(0.5, 0.5, 'Cannot read attack image', ha='center', va='center')
        else:
            plt.text(0.5, 0.5, 'No attack image', ha='center', va='center')
        plt.axis('off')

        plt.subplot(n_pairs, 2, i * 2 + 2)
        if i < len(recover_images):
            img_r = cv2.imread(str(recover_images[i]), cv2.IMREAD_GRAYSCALE)
            if img_r is not None:
                plt.imshow(img_r, cmap='gray')
                plt.title(f'Recover: {recover_images[i].name}')
            else:
                plt.text(0.5, 0.5, 'Cannot read recover image', ha='center', va='center')
        else:
            plt.text(0.5, 0.5, 'No recover image', ha='center', va='center')
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```
