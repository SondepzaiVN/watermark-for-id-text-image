```python
from pathlib import Path
import cv2
import matplotlib.pyplot as plt

ROOT = Path('..').resolve()
host_dir = ROOT / 'data' / 'host_images'
wm_dir = ROOT / 'data' / 'watermark'

print('Project root:', ROOT)
print('Host dir:', host_dir)
print('Watermark dir:', wm_dir)
```

```python
def list_images(folder: Path):
    exts = ['*.png', '*.jpg', '*.jpeg', '*.bmp']
    files = []
    for ext in exts:
        files.extend(folder.glob(ext))
    return sorted(files)

host_images = list_images(host_dir)
watermark_images = list_images(wm_dir)

print('So host images:', len(host_images))
for p in host_images[:10]:
    print(' -', p.name)

print('So watermark images:', len(watermark_images))
for p in watermark_images[:10]:
    print(' -', p.name)
```

```python
def read_rgb(path: Path):
    img_bgr = cv2.imread(str(path), cv2.IMREAD_COLOR)
    if img_bgr is None:
        return None
    return cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

n_show = 6
sample_hosts = host_images[:n_show]

if not sample_hosts:
    print('Khong tim thay host image. Hay them anh vao data/host_images/.')
else:
    cols = 3
    rows = (len(sample_hosts) + cols - 1) // cols
    plt.figure(figsize=(14, 4 * rows))
    for i, p in enumerate(sample_hosts, start=1):
        img = read_rgb(p)
        plt.subplot(rows, cols, i)
        if img is None:
            plt.text(0.5, 0.5, f'Cannot read\n{p.name}', ha='center', va='center')
            plt.axis('off')
            continue
        plt.imshow(img)
        plt.title(p.name)
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
if not watermark_images:
    print('Khong tim thay watermark image. Hay them vao data/watermark/.')
else:
    plt.figure(figsize=(10, 4))
    for i, p in enumerate(watermark_images[:4], start=1):
        img = cv2.imread(str(p), cv2.IMREAD_GRAYSCALE)
        plt.subplot(1, min(4, len(watermark_images)), i)
        if img is None:
            plt.text(0.5, 0.5, f'Cannot read\n{p.name}', ha='center', va='center')
            plt.axis('off')
            continue
        plt.imshow(img, cmap='gray')
        plt.title(p.name)
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```
