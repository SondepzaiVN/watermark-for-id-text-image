# 04 - Metrics And Charts
Notebook nay doc file Excel tong hop va ve bieu do PSNR/SSIM/NC/BER.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path('..').resolve()
excel_path = ROOT / 'results' / 'watermark_evaluation_summary.xlsx'
print('Excel path:', excel_path)
print('Exists:', excel_path.exists())
```

```python
if not excel_path.exists():
    raise FileNotFoundError('Khong tim thay watermark_evaluation_summary.xlsx. Hay chay scripts/main.py truoc.')

xls = pd.ExcelFile(excel_path)
sheet_names = xls.sheet_names
print('So luong sheets:', len(sheet_names))
print('Sheets:', sheet_names[:20])
```

```python
frames = []
for name in sheet_names:
    df = pd.read_excel(excel_path, sheet_name=name)
    df['host'] = name
    frames.append(df)

all_df = pd.concat(frames, ignore_index=True)
print('Tong so dong:', len(all_df))
all_df.head()
```

```python
candidate_cols = ['PSNR', 'SSIM', 'NC', 'BER']
metric_cols = [c for c in candidate_cols if c in all_df.columns]

if not metric_cols:
    raise ValueError('Khong tim thay cot metric nao trong file Excel.')

summary = all_df[metric_cols].mean(numeric_only=True).to_frame('mean').reset_index().rename(columns={'index': 'metric'})
summary
```

```python
plt.figure(figsize=(8, 4))
plt.bar(summary['metric'], summary['mean'])
plt.title('Mean Metrics Across All Hosts And Attacks')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
for col in metric_cols:
    plt.figure(figsize=(10, 4))
    all_df[col].plot(kind='hist', bins=20)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.grid(axis='y', linestyle='--', alpha=0.3)
    plt.tight_layout()
    plt.show()
```
